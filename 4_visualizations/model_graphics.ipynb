```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### Visualizing  Classes

```python
# glucose
classes = ('0', '1')
interval = np.arange(len(classes))
frequencies = [4746, 361]

fig = plt.figure()
plt.bar(interval, frequencies, color='#A9678F')
plt.xticks(interval, classes)
plt.ylabel('Frequencies')
plt.title('Glucose: C_JOINT Distribution')

plt.show()
fig.savefig('gluc_distribution.png')
```

```python
# inflammation
classes = ('1', '2', '3')
interval = np.arange(len(classes))
frequencies = [1870, 1371, 1298]

fig = plt.figure()
plt.bar(interval, frequencies, color='#2D1E3E')
plt.xticks(interval, classes)
plt.ylabel('Frequencies')
plt.title('Inflammation: C_CRP Distribution')

plt.show()
fig.savefig('imm_distribution.png')
```

```python
# lipids
classes = ('0', '1')
interval = np.arange(len(classes))
frequencies = [4686, 421]

fig = plt.figure()
plt.bar(interval, frequencies, color='#DDA9AF')
plt.xticks(interval, classes)
plt.ylabel('Frequencies')
plt.title('Lipids: C_JOINT2 Distribution')

plt.show()
fig.savefig('lip_distribution.png')
```

#### Comparing glucose models

```python
# Init Dec Tree, Init Random Forest, Over DT, Over RF
precision = [0.17, 0.11, 0.58, 0.61]
recall = [0.11, 0.43, 0.47, 0.54]
counts = len(precision)
```

```python
figure = plt.figure()

x = np.arange(counts)
width = 0.35

figure, ax = plt.subplots()
bar1 = ax.bar(x, precision, width, color='#DDA9AF')
bar2 = ax.bar(x + width, recall, width, color='#2D1E3E')

ax.set_ylabel('Score')
ax.set_ybound(upper=1)
ax.set_title('Performance Metrics for Glucose Models (C_JOINT=1)')
ax.set_xticks(x + width)
ax.set_xticklabels(('DT', 'RF (weighted)', 'OS-DT', 'OS-RF'))
ax.legend((bar1[0], bar2[0]), ('Precision', 'Recall'))

plt.show()
figure.savefig('glucose_performance.png')
```

#### Comparing best two and three class models

```python
precision = [0.45, 0.62]
recall = [0.83, 0.76]
f1 = [0.58, 0.68]
counts = len(precision)
```

```python
figure = plt.figure()

x = np.arange(counts)
width = 0.28

figure, ax = plt.subplots()
bar1 = ax.bar(x, precision, width, color='#DDA9AF')
bar2 = ax.bar(x + width, recall, width, color='#2D1E3E')
bar3 = ax.bar(x + 2*width, f1, width, color='#A9678F')

ax.set_ylabel('Score')
ax.set_ybound(upper=1)
ax.set_title('Inflammation: Best 2-Class vs Best 3-Class (C_CRP=3)')
ax.set_xticks(x + width)
ax.set_xticklabels(('3 Class RF', '2 Class OS-RF'))
ax.legend((bar1[0], bar2[0], bar3[0]), ('Precision', 'Recall', 'F1-Score'))

plt.show()
figure.savefig('imm_comparison_performance.png')
```
